## **Necessary Imports**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df.info()
```

```python
df.sample(10)
```

*Converting the data type as the data was all numeric*

```python
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')
```

```python
df['Rating'].unique()
```

*Checking null values*

```python
df.isna().sum()
```

*Using mean and mode to fill the null values*

```python
for i in df.select_dtypes(include='number'):
    df[i] = df[i].fillna(df[i].mean())

for i in df.select_dtypes(include='object'):
    df[i] = df[i].fillna(df[i].mode()[0])


df.isna().sum()
```

*Scaling the data for outlier detection*

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df.shape
```

*Using one class support vector machine to remove outliers*

```python
from sklearn.svm import OneClassSVM

ocsvm = OneClassSVM(nu=0.05,kernel='rbf',gamma='auto')
scaled_features = scaler.fit_transform(df.select_dtypes(include='number'))
outliers_svm = ocsvm.fit_predict(scaled_features)
df['Outlier_Svm'] = outliers_svm
df = df.loc[df['Outlier_Svm'] == 1]
```

```python
df = df.drop(columns='Outlier_Svm')
```

*Outliers removed*

```python
df.shape
```

## **Feature analysis and Model Training**

*Name was too unique for encoding*

```python
encoded_df = df.drop(columns='Name')
```

*Label encoding for categorical columns*

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in encoded_df.select_dtypes(include='object'):
    encoded_df[i] = le.fit_transform(encoded_df[i])

encoded_df.shape
```

```python
encoded_df.sample(10)
```

```python
encoded_df.info()
```

**Stripped down multiple features by tracking the accuracy of the model using Lazpredict**

```python
X = encoded_df.drop(columns=['Global_Sales','EU_Sales','JP_Sales','Other_Sales','Critic_Score','Critic_Count','User_Score','User_Count','Developer','Rating'])
y = encoded_df['Global_Sales']
```

*Columns that the model is being trained on*

```python
for i in df.columns:
    if i in X.columns:
        print(i)
```

```python
X_scale = scaler.fit_transform(X)
```

*Forming small subsets of the data so lazypredict can give quicker responses*

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scale, y, test_size=0.2, random_state=30)

X_small = X_train[:2000]
y_small = y_train[:2000]
X_small_test = X_test[:500]
y_small_test = y_test[:500]
```

```python
from lazypredict.Supervised import LazyRegressor

lzc = LazyRegressor(verbose=0, ignore_warnings=True)
models, predictions = lzc.fit(X_small, X_small_test, y_small, y_small_test)
```

```python
models
```

*Using lazypredict as it gives the best accuracy*

```python
from sklearn.neural_network import MLPRegressor
model = MLPRegressor()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```

*Cheking the metrics of the trained model*

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R Score: {r2}')
```
